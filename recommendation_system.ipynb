```python
from sklearn.model_selection import train_test_split

def train_validation_test_split(Y, R,  val_size=0.2, test_size=0.2):
    """
    Splits the ratings matrix Y into training, validation, and test sets.

    Parameters:
    - Y: Ratings matrix (m x n)
    - R: Indicator matrix, where R(i, j) = 1 if user j rated movie i
    - test_size: Proportion of the data to be used as the test set
    - val_size: Proportion of the remaining data to be used as the validation set

    Returns:
    - Y_train, Y_val, Y_test: The split data matrices
    - R_train, R_val, R_test: Corresponding indicator matrices
    """
    Y_geral = Y.copy().values
    # Get indices of non-zero entries in the ratings matrix Y
    movie_idx, user_idx = np.where(R == 1)
    indices = list(zip(movie_idx, user_idx))

    # Split into train + (validation + test)
    train_indices, test_indices = train_test_split(indices, test_size=test_size, random_state=42)

    # Further split train into train + validation
    train_indices, val_indices = train_test_split(train_indices, test_size=val_size, random_state=42)

    # Initialize train, val, and test sets with zeros
    Y_train = np.zeros(Y_geral.shape)
    Y_val = np.zeros(Y_geral.shape)
    Y_test = np.zeros(Y_geral.shape)

    R_train = np.zeros(R.shape)
    R_val = np.zeros(R.shape)
    R_test = np.zeros(R.shape)

    # Populate the training set
    for i, j in train_indices:
        Y_train[i, j] = Y_geral[i, j]
        R_train[i, j] = 1

    # Populate the validation set
    for i, j in val_indices:
        Y_val[i, j] = Y_geral[i, j]
        R_val[i, j] = 1

    # Populate the test set
    for i, j in test_indices:
        Y_test[i, j] = Y_geral[i, j]
        R_test[i, j] = 1

    return Y_train, R_train, Y_val, R_val, Y_test, R_test
```

```python
def normalize_ratings(Y_train, R_train):
    """
    Normalize the ratings matrix Y_train by subtracting the mean rating for each movie.

    Parameters:
    - Y_train: Training ratings matrix (m x n)
    - R_train: Indicator matrix for training data (m x n)

    Returns:
    - Y_train_norm: Normalized training ratings matrix
    - Y_train_mean: Mean ratings for each movie
    """
    # Calculate the mean for each movie only where there is a rating
    Y_train_mean = np.where(R_train.sum(axis=1) > 0,
                             Y_train.sum(axis=1) / R_train.sum(axis=1),
                             0)  # Average rating per movie

    return Y_train_mean

# Split the data
Y_train, R_train, Y_val, R_val, Y_test, R_test = train_validation_test_split(Y, R, val_size=0.20, test_size=0.10)

# Normalize the training ratings by film (rows)
Y_train_mean = normalize_ratings(Y_train, R_train)
```

```python
# Normalize the training set
Y_train_norm = Y_train - Y_train_mean[:, np.newaxis]

Y_val_norm = Y_val - Y_train_mean[:, np.newaxis]

Y_test_norm = Y_test - Y_train_mean[:, np.newaxis]
```

```python
def cofi_cost_func_v(X, W, b, Y, R, lambda_):
    """
    Returns the cost for the content-based filtering
    Vectorized for speed. Uses tensorflow operations to be compatible with custom training loop.
    Args:
      X (ndarray (num_movies,num_features)): matrix of item features
      W (ndarray (num_users,num_features)) : matrix of user parameters
      b (ndarray (1, num_users)            : vector of user parameters
      Y (ndarray (num_movies,num_users)    : matrix of user ratings of movies
      R (ndarray (num_movies,num_users)    : matrix, where R(i, j) = 1 if the i-th movies was rated by the j-th user
      lambda_ (float): regularization parameter
    Returns:
      J (float) : Cost
    """
    j = (tf.linalg.matmul(X, tf.transpose(W)) + b - Y)*R
    J = 0.5 * tf.reduce_sum(j**2) + (lambda_/2) * (tf.reduce_sum(X**2) + tf.reduce_sum(W**2))
    return J
```

```python
import optuna
import tensorflow as tf
from tensorflow import keras
# Hiperparameter optimization
def objective(trial):
    num_features = trial.suggest_categorical('num_features', [1, 3, 5, 7, 10, 30, 60, 90, 120])
    lambda_ = trial.suggest_categorical('lambda_', [0, 0.001, 0.002,0.003,0.005, 0.01, 0.05, 0.1, 0.5, 1,1.5, 2,2.5,3])
    learning_rate = trial.suggest_categorical('learning_rate', [0.001, 0.005, 0.008, 0.01, 0.05, 0.1])

    num_movies, num_users = Y_train.shape

    # Set Initial Parameters (W, X)
    tf.random.set_seed(1234)  # for consistent results
    W = tf.Variable(tf.random.normal((num_users, int(num_features)), dtype=tf.float64), name='W')
    X = tf.Variable(tf.random.normal((num_movies, int(num_features)), dtype=tf.float64), name='X')
    b = tf.Variable(tf.random.normal((1, num_users), dtype=tf.float64), name='b')

    # Instantiate an optimizer with the given learning rate
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)

    iterations = 300
    for iter in range(iterations):
        with tf.GradientTape() as tape:
            cost_value = cofi_cost_func_v(X, W, b, Y_train_norm, R_train, lambda_)

        grads = tape.gradient(cost_value, [X, W, b])
        optimizer.apply_gradients(zip(grads, [X, W, b]))

        # Log validation loss at the end of training
        if iter == iterations - 1:
            # train_cost_value = cost_value
            val_cost_value = cofi_cost_func_v(X, W, b, Y_val_norm, R_val, lambda_)
            return val_cost_value #+ abs(val_cost_value - train_cost_value)  # Return validation loss for optimization

# Create a study object and optimize the objective function
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=15)

# Print the best results
print("Best hyperparameters: ", study.best_params)
print("Best validation loss: ", study.best_value)
```

```python
def compute_rmse_mae(Y_true, Y_pred, R):
    """
    Compute the RMSE and MAE between the true and predicted ratings.

    Parameters:
    - Y_true: True ratings matrix (m x n)
    - Y_pred: Predicted ratings matrix (m x n)
    - R: Indicator matrix, where R(i, j) = 1 if user j rated movie i

    Returns:
    - RMSE: Root Mean Squared Error
    - MAE: Mean Absolute Error
    """
    # Only consider predicted ratings where R == 1
    error = (Y_pred - Y_true) * R
    squared_error = error ** 2

    # RMSE
    mse = np.sum(squared_error) / np.sum(R)
    rmse = np.sqrt(mse)

    # MAE
    mae = np.sum(np.abs(error)) / np.sum(R)

    return rmse, mae
```

```python
# Final model to test the metrics
num_movies, num_users = Y_train.shape
num_features = study.best_params['num_features']

# Set Initial Parameters (W, X), use tf.Variable to track these variables
tf.random.set_seed(1234) # for consistent results
W = tf.Variable(tf.random.normal((num_users,  num_features),dtype=tf.float64),  name='W')
X = tf.Variable(tf.random.normal((num_movies, num_features),dtype=tf.float64),  name='X')
b = tf.Variable(tf.random.normal((1,num_users),   dtype=tf.float64),  name='b')

# Instantiate an optimizer.
optimizer = keras.optimizers.Adam(learning_rate=study.best_params['learning_rate'])

iterations = 500
lambda_ = study.best_params['lambda_']

# Define early stopping parameters
patience = 10  # Number of iterations to wait before stopping after last improvement
best_val_rmse = float('inf')  # Initialize with a large value
wait = 0  # Counter for patience

for iter in range(iterations):
    # Use TensorFlows GradientTape

    with tf.GradientTape() as tape:

        # Compute the cost (forward pass included in cost)
        cost_value = cofi_cost_func_v(X, W, b, Y_train_norm, R_train, lambda_)

    # Calcula as derivadas parciais de cada varivel
    # Utiliza o tf.Variable para identificar as derivadas parciais que deve fazer
    grads = tape.gradient( cost_value, [X,W,b] )

    # Run one step of gradient descent by updating
    # the value of the variables to minimize the loss.
    optimizer.apply_gradients( zip(grads, [X,W,b]) )

    # Calculate validation cost
    val_cost_value = cofi_cost_func_v(X, W, b, Y_val_norm, R_val, lambda_)
    Y_pred = np.dot(X.numpy(), W.numpy().T) + b.numpy() + Y_train_mean[:, np.newaxis]
    # 2. Calculate RMSE and MAE for training set
    train_rmse, train_mae = compute_rmse_mae(Y_train, Y_pred, R_train)
    # 3. Calculate RMSE and MAE for validation set
    val_rmse, val_mae = compute_rmse_mae(Y_val, Y_pred, R_val)
    # Check for early stopping
    if val_rmse < best_val_rmse:
        best_val_rmse = val_rmse  # Update the best RMSE
        wait = 0  # Reset patience counter
    else:
        wait += 1  # Increment patience counter

    # Stop training if the validation RMSE hasn't improved in the last 'patience' iterations
    if wait >= patience:
        print(f"Early stopping at iteration {iter} with best validation RMSE: {best_val_rmse:.4f}")
        break  # Stop the training loop

    # Print periodically.
    if (iter % 50 == 0) or (iter % 499 == 0):
        print(f"Training loss at iteration {iter}: {cost_value:0.1f}")
        print(f"Validation loss at iteration {iter}: {val_cost_value:0.1f}")
        print()
        print(f"Train RMSE: {train_rmse:.4f}, Train MAE: {train_mae:.4f}")
        print(f"Validation RMSE: {val_rmse:.4f}, Validation MAE: {val_mae:.4f}")
        print()

# Calculate final error term.
test_rmse, teste_mae = compute_rmse_mae(Y_test, Y_pred, R_test)
print(f"Final Test RMSE: {test_rmse:.4f}, Final test MAE: {teste_mae:.4f}")
```
